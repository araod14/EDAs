```python
nba_df.Height.str.split("-", expand=True)
```

```python
H_feet=nba_df.Height.str.split("-",expand=True)[0] 
H_inch=nba_df.Height.str.split("-",expand=True)[1]

nba_df["Heigt_ft"]=pd.to_numeric(H_feet, errors="coerce") + pd.to_numeric(H_inch, errors="coerce")*0.083
```

```python
nba_df.head()
```

```python
nba_df["Height_m"]=nba_df["Heigt_ft"] / 3.28
```

```python
nba_df.head()
```

```python
nba_df.Height_m.describe()
```

```python
nba_df["BMI"] = (nba_df.Weight / 2.2) / (nba_df.Height_m)**2
```
